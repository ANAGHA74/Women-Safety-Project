```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(df.columns)
```

```python
df = df.drop(columns=["date"])
```

```python
df.head()
```

```python
df.shape
```

```python
df["time_clean"] = (
    df["time"]
    .astype(str)
    .str.lower()
    .str.replace("hrs", "", regex=False)
    .str.replace("hour", "", regex=False)
    .str.strip()
)
```

```python
df["hour"] = pd.to_datetime(
    df["time_clean"],
    errors="coerce"
).dt.hour
```

```python
df["hour"] = pd.to_datetime(df["time_clean"],errors="coerce").dt.hour
```

```python
print("Invalid hour values:", df["hour"].isna().sum())

df = df.dropna(subset=["hour"])
```

```python
df.shape
```

```python
df.head()
```

```python
df["is_night"] = df["hour"].apply(
    lambda x: 1 if x >= 19 or x <= 5 else 0
)
```

```python
df["latitude"] = (
    df["latitude"]
    .astype(str)
    .str.strip()
    .str.replace(",", "")
)

df["longitude"] = (
    df["longitude"]
    .astype(str)
    .str.strip()
    .str.replace(",", "")
)

df["latitude"] = pd.to_numeric(df["latitude"], errors="coerce")
df["longitude"] = pd.to_numeric(df["longitude"], errors="coerce")
```

```python
df = df.dropna(
    subset=["year", "latitude", "longitude"]
)
```

```python
df=df.drop_duplicates()
```
